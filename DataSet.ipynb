```python
def combine_images_horizontal(img1, img2):
    # Assuming img1 and img2 are PIL Image objects
    combined_img = Image.new('RGB', (img1.width + img2.width, img1.height))
    combined_img.paste(img1, (0, 0))
    combined_img.paste(img2, (img1.width, 0))
    return combined_img
```

```python
def split_image_horizontal(combined_img):
    # Assuming combined_img is a PIL Image object
    img1 = combined_img.crop((0, 0, combined_img.width // 2, combined_img.height))
    img2 = combined_img.crop((combined_img.width // 2, 0, combined_img.width, combined_img.height))
    return img1, img2
```
